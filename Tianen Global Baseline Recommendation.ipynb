### Read Original Dataset with only Answers

```python
import pandas as pd
import numpy as np
df = pd.read_csv('https://raw.githubusercontent.com/fbhsaid/DAV6300/main/Global%20Baseline%20Data.csv')
df.head()
```

### Reshape Data and Add UserAvg Column and PhoneAvg Row

```python
df.index = ['Olivia', 'ZJY', 'LHR', 'ZHC','ZYC','Tianen']
df.columns =['Col_1', 'Huawei', 'Xiaomi', 'Vivo', 'Apple', 'Samsung', 'Sony']
df1 = df[['Huawei', 'Xiaomi', 'Vivo', 'Apple', 'Samsung', 'Sony']]
df1.loc[-1] = np.nan
df1.index = ['Olivia', 'ZJY', 'LHR', 'ZHC','ZYC','Tianen', 'PhoneAvg']
df1['UserAvg'] = np.nan
df1
```

### Calculate MeanPhone

```python
df1.at['PhoneAvg', 'UserAvg'] = (df1['Huawei'].sum() + df1['Xiaomi'].sum() + df1['Vivo'].sum() + df1['Apple'].sum() + df1['Samsung'].sum() + df1['Sony'].sum()) / (df1['Huawei'].notnull().sum() + df1['Xiaomi'].notnull().sum() + df1['Vivo'].notnull().sum() + df1['Apple'].notnull().sum() + df1['Samsung'].notnull().sum() + df1['Sony'].notnull().sum())
df1
```

### Calculate PhoneAvg

```python
df1.at['PhoneAvg', 'Huawei'] = df1['Huawei'].sum() / df1['Huawei'].notnull().sum()
df1.at['PhoneAvg', 'Xiaomi'] = df1['Xiaomi'].sum() / df1['Xiaomi'].notnull().sum()
df1.at['PhoneAvg', 'Vivo'] = df1['Vivo'].sum() / df1['Vivo'].notnull().sum()
df1.at['PhoneAvg', 'Apple'] = df1['Apple'].sum() / df1['Apple'].notnull().sum()
df1.at['PhoneAvg', 'Samsung'] = df1['Samsung'].sum() / df1['Samsung'].notnull().sum()
df1.at['PhoneAvg', 'Sony'] = df1['Sony'].sum() / df1['Sony'].notnull().sum()
df1
```

### Calculate UserAvg

```python
df1.at['Olivia', 'UserAvg'] = df1.iloc[0].sum() / df1.iloc[0].notnull().sum()
df1.at['ZJY', 'UserAvg'] = df1.iloc[1].sum() / df1.iloc[1].notnull().sum()
df1.at['LHR', 'UserAvg'] = df1.iloc[2].sum() / df1.iloc[2].notnull().sum()
df1.at['ZHC', 'UserAvg'] = df1.iloc[3].sum() / df1.iloc[3].notnull().sum()
df1.at['ZYC', 'UserAvg'] = df1.iloc[4].sum() / df1.iloc[4].notnull().sum()
df1.at['Tianen', 'UserAvg'] = df1.iloc[5].sum() / df1.iloc[5].notnull().sum()
df1
```

### Add PhoneAvg - MeanPhone Row

```python
df1.loc[-1] = np.nan
df1.index = ['Olivia', 'ZJY', 'LHR', 'ZHC','ZYC','Tianen', 'PhoneAvg', 'PhoneAvg - MeanPhone']
df1
```

### Calculate PhoneAvg - MeanPhone Row

```python
df1.at['PhoneAvg - MeanPhone', 'Huawei'] = df1.at['PhoneAvg', 'Huawei'] - df1.at['PhoneAvg', 'UserAvg']
df1.at['PhoneAvg - MeanPhone', 'Xiaomi'] = df1.at['PhoneAvg', 'Xiaomi'] - df1.at['PhoneAvg', 'UserAvg']
df1.at['PhoneAvg - MeanPhone', 'Vivo'] = df1.at['PhoneAvg', 'Vivo'] - df1.at['PhoneAvg', 'UserAvg']
df1.at['PhoneAvg - MeanPhone', 'Apple'] = df1.at['PhoneAvg', 'Apple'] - df1.at['PhoneAvg', 'UserAvg']
df1.at['PhoneAvg - MeanPhone', 'Samsung'] = df1.at['PhoneAvg', 'Samsung'] - df1.at['PhoneAvg', 'UserAvg']
df1.at['PhoneAvg - MeanPhone', 'Sony'] = df1.at['PhoneAvg', 'Sony'] - df1.at['PhoneAvg', 'UserAvg']
df1
```

### Add UserAvg - MeanPhone Column

```python
df1['UserAvg - MeanPhone'] = np.nan
df1
```

### Calculate UserAvg - MeanPhone Column

```python
df1.at['Olivia', 'UserAvg - MeanPhone'] = df1.at['Olivia', 'UserAvg'] - df1.at['PhoneAvg', 'UserAvg']
df1.at['ZJY', 'UserAvg - MeanPhone'] = df1.at['ZJY', 'UserAvg'] - df1.at['PhoneAvg', 'UserAvg']
df1.at['LHR', 'UserAvg - MeanPhone'] = df1.at['LHR', 'UserAvg'] - df1.at['PhoneAvg', 'UserAvg']
df1.at['ZHC', 'UserAvg - MeanPhone'] = df1.at['ZHC', 'UserAvg'] - df1.at['PhoneAvg', 'UserAvg']
df1.at['ZYC', 'UserAvg - MeanPhone'] = df1.at['ZYC', 'UserAvg'] - df1.at['PhoneAvg', 'UserAvg']
df1.at['Tianen', 'UserAvg - MeanPhone'] = df1.at['Tianen', 'UserAvg'] - df1.at['PhoneAvg', 'UserAvg']
df1
```
